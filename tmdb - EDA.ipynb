```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
year2000_df = pd.read_csv("Data/final_tmdb_data_2000.csv.gz")
year2001_df = pd.read_csv("Data/final_tmdb_data_2001.csv.gz")
```

```python
year2000_df
```

```python
year2001_df
```

```python
combined_df = year2000_df.append(year2001_df)
combined_df
```

How many movies had at least some valid financial information?
(values > 0 for budget OR revenue)

```python
financial_filter = (combined_df['budget'] > 0) | (combined_df['revenue'] > 0)
combined_financial_df = combined_df[financial_filter]
combined_financial_df.count()

#There are 940 movies with relvant financial data
```

How Many movies are there in each of the certification categories (G/PG/PG-13/R)?


```python
combined_financial_df['certification'].value_counts()
```

```python
revenue_mean_df = combined_financial_df.groupby('certification')\
            ['revenue']
revenue_mean_df.mean()
```

What is the average budget per certification category?

```python
budget_mean_df = combined_financial_df.groupby('certification')\
            ['budget']
budget_mean_df.mean()
```
